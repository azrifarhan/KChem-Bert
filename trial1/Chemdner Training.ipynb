```python
def convert_to_spacy_format(df):
    result = []

    for _, row in df.iterrows():
        text = row['text']
        tokens = row['tokens']
        tags = row['labels']

        entities = []
        current_pos = 0
        entity = None

        for token, label in zip(tokens, tags):
            try:
                start = text.index(token, current_pos)
                end = start + len(token)
                current_pos = end

                if label.startswith("B-"):
                    if entity:
                        entities.append(tuple(entity))
                    entity = [start, end, "Chemical"]

                elif label.startswith("I-") and entity:
                    entity[1] = end

                else:  # label == "O"
                    if entity:
                        entities.append(tuple(entity))
                        entity = None

            except ValueError:
                continue 
        if entity:
            entities.append(tuple(entity))

        result.append({'text': text, "entities": entities})

    return result
```

```python
from spacy.tokens import DocBin
from tqdm import tqdm
import spacy
from spacy.util import filter_spans
import json
```

```python
nlp = spacy.blank('en')
doc_bin = DocBin()

for lmao in tqdm(test):
    text = lmao['text']
    labels = lmao['entities']
    doc = nlp.make_doc(text)
    ents = []

    for start, end, label in labels:
        span = doc.char_span(start, end, label = label, alignment_mode="contract")
        if span is None:
            print("Skip")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents
    doc_bin.add(doc)

doc_bin.to_disk("dev.spacy")


for lmao in tqdm(train):
    text = lmao['text']
    labels = lmao['entities']
    doc = nlp.make_doc(text)
    ents = []

    for start, end, label in labels:
        span = doc.char_span(start, end, label = label, alignment_mode="contract")
        if span is None:
            print("Skip")
        else:
            ents.append(span)
    filtered_ents = filter_spans(ents)
    doc.ents = filtered_ents
    doc_bin.add(doc)

doc_bin.to_disk("train.spacy")
    
```
